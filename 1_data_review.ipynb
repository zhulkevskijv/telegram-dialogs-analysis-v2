7. Calculate top-10 people to whom you wrote the biggest amount of messages (name, amount of messages).

```python
my_messages = df.loc[df["from_id"] == f"PeerUser(user_id={telegram_id})"]
```

```python
private_dialogs_ids = df_meta[df_meta["type"] == "Private dialog"]["dialog_id"]
```

```python
sent_messages_private_dialogs = my_messages[my_messages["dialog_id"].isin(private_dialogs_ids)]
sent_messages_top_10 = sent_messages_private_dialogs[["id","dialog_id"]].groupby(by="dialog_id", as_index=False).count().sort_values("id", ascending=False).head(10)
```

```python
import ast
```

```python
df_meta_names = df_meta.copy()
df_meta_names["user_id"] = df_meta_names["users"].apply(lambda x: ast.literal_eval(x)["user_id"])
df_meta_names["user_name"] = df_meta_names["users"].apply(lambda x: str(ast.literal_eval(x)["first_name"]))
df_meta_names
```

```python
sent_messages_top_10["name"] = sent_messages_top_10["dialog_id"].apply(lambda x: df_meta_names[df_meta_names["dialog_id"] == x]["user_name"].iloc[0])
```

```python
sent_messages_top_10
```
